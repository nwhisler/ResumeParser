```python
from pypdf import PdfReader
```

```python
reader = PdfReader("Nicholas_Whisler_Generic_Resume_Timeline.pdf")
```

```python
text = ""
for page in reader.pages:
    text += page.extract_text() + "\n"
```

```python
prompt = f"""
            You are processing resumes to extract the name of the candidate, their address, educational experience, work experience, academic projects, and personal projects.
            {text}
          """
```

```python
from pydantic import BaseModel
```

```python
class DataExtraction(BaseModel):
  name:str
  address:str
  education:str
  work:str
  academic_projects:str
  personal_projects:str
```

```python
import json
```

```python
response = ollama.chat(model='llama3', messages=[
    {
        'role': 'user',
        'content': prompt,
    },
], format=DataExtraction.model_json_schema())
data = response['message']['content']
```

```python
json_data = json.loads(data)
```
